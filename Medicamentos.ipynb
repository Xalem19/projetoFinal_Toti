```python
# Importamos a tabela de Presc_pro

Presc_pro = pd.read_csv('/content/Consul_pres_pro.csv',encoding='latin-1')

Presc_pro
```

```python
#para verificar o tipo de dados
Presc_pro.dtypes
```

```python
#verificando mudança de tipo de dado
Presc_proData=Presc_pro['Data'].astype(np.datetime64)
Presc_proData
```

```python
#codigo utilizado para modificar a formato data, por só mẽs.
Presc_pro['Data']= Presc_proData.dt.month_name()
Presc_pro
```

```python
#Se observo que existen dados duplicados.  
Presc_pro.duplicated().sum()
```

```python
#Para corregir os dados duplicados se utilizo a siguente função
Presc_pro = Presc_pro.drop_duplicates()
Presc_pro
```

```python
#Exportando tabela a csv
Presc_pro.to_csv('Prescpro_BI1.csv')
```
